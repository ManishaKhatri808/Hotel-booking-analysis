```python
#Data visualization and analysis
```

```python
is_canceled=df['is_canceled'].value_counts(normalize=True)
print(is_canceled)

plt.figure(figsize=(3,3))
plt.title('Reservation status counts')
plt.bar(['Not canceled','canceled'],df['is_canceled'].value_counts())
```

```python
plt.figure(figsize=(4,4))
ax1=sn.countplot(x='hotel',hue='is_canceled',data=df)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
df
```

```python
resort_hotel=resort_hotel.groupby(['reservation_status_date'])[['adr']].mean()
resort_hotel
```

```python
city_hotel=city_hotel.groupby(['reservation_status_date'])[['adr']].mean()
city_hotel
```

```python
plt.figure(figsize=(20,10))
plt.title('Average price rate of city and resort hotels',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='resort hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='city hotel')
plt.legend()
plt.show()
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df['month']=df['reservation_status_date'].dt.month
```

```python
plt.figure(figsize=(10,5))
ax1=sn.countplot(x='month',hue='is_canceled',data=df)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not_canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sn.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
```

```python
canceled_data=df[df['is_canceled']==1]
top_ten_countries=canceled_data['country'].value_counts()[:10]
```

```python
plt.figure(figsize=(5,5))
plt.title('Top Ten Countries with reservation cancelation')
plt.pie(top_ten_countries,autopct='%.2f',labels=top_ten_countries.index)
```

```python
canceled_data_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_data_adr.sort_values('reservation_status_date',inplace=True)
canceled_data_adr.reset_index(inplace=True)
```
